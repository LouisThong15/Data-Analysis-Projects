## Data Cleaning and Preparation

Firstly, we will remove some columns that are irrelevant for our analysis such as 'Address', 'SellerG', 'Longtitude' and 'Lattitude'.

```python
# Dropping unnecessary columns
df.drop(columns = ['Address', 'Longtitude', 'Lattitude', 'SellerG'], inplace = True)
```

We are also going to rename the `Method` column to `Status` for easier understanding.

```python
# Renaming column names
df.rename(columns = {'Method':'Status'}, inplace = True)
```

### Dealing with null values

There are null values in the `Car`, `BuildingArea`, `YearBuilt` and `CouncilArea` column. We will replace these values with their respective median value, except for `CouncilArea`. For `CouncilArea`, we will use the mode value. 

```python
# Filling null values with median and mode value
df['Car'].fillna(df['Car'].median(), inplace = True)
df['BuildingArea'].fillna(df['BuildingArea'].median(), inplace = True)
df['YearBuilt'].fillna(df['YearBuilt'].median(), inplace = True)
df['CouncilArea'].fillna(df['CouncilArea'].mode()[0], inplace = True)
```

### Removing outliers

We will remove the extreme outliers in `BuildingArea` and `Landsize` by using the IQR method.

```python
# IQR method to remove extreme outliers
for col in ['BuildingArea', 'Landsize']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

In the dataset, there is a property which states that it was built in the year 1196. We will consider this as an outlier and remove it. 

```python
# Initial YearBuilt count
df.YearBuilt.value_counts().rename_axis('Year').reset_index().sort_values(by = 'Year').head()
```

```python
# Filtering YearBuilt
df = df[df.YearBuilt > 1800]
```

### Converting data types

We will convert the data type of the `Date` column from object to datetime.

```python
# Converting Date column to datetime
df['Date'] = pd.to_datetime(df.Date, format ='%d/%m/%Y')
```

```python
# Plotting boxplot for price against property type
sns.boxplot(data = modified_df, y = 'Price', x = 'Type', palette = 'hls', showfliers = False)
plt.title('Boxplot of Property Types and Log Price')
plt.show()

# Printing the median price for property types
print('Median price of houses: $' +'%.0f' % np.median(df[df.Type == 'h'].Price))
print('Median price of units: $' + '%.0f' % np.median(df[df.Type == 'u'].Price))
print('Median price of townhouses: $'+ '%.0f' % np.median(df[df.Type == 't'].Price))
```

It seems that the median price for each property type is quite different. We will continue to perform ANOVA to determine if there is a significant difference. 

#### Analysis of Variance (ANOVA)

```python
# OLS
ols_formula = 'log_price ~ C(Type)'
model = ols(formula = ols_formula, data = modified_df).fit()
model.summary()
```

#### Null Hypothesis

$$H_0: price_h=price_t=price_u$$

There is a no difference in the price of properties based on property types.

#### Alternative Hypothesis

$$H_1: \text{Not } price_h=price_t=price_u$$

```python
sm.stats.anova_lm(model, typ = 1)
```

```python
sm.stats.anova_lm(model, typ = 2)
```

```python
sm.stats.anova_lm(model, typ = 3)
```

Since all p-values are less than the significance threshold of 0.5, we can reject the null hypothesis and accept the alternative hypothesis. Next, we will use Tukey One Way ANOVA for further analysis to determine if there is significant difference between each property type and their prices. 

#### Tukey One Way ANOVA

```python
# Performing Tukey One Way ANOVA with Type
from statsmodels.stats.multicomp import pairwise_tukeyhsd
tukey_oneway = pairwise_tukeyhsd(endog = modified_df['log_price'], groups = modified_df['Type'])
tukey_oneway.summary()
```

From the results above, it can be determined that there is a significant difference between the prices of each property type since the reject column is `True` for all. 

## Prediction of Property Prices

### Feature Engineering

```python
# Creating dependent and independent variables
X = df.drop(columns = {'log_price', 'Date', 'Price', 'Bedroom2'})
y = df.log_price
```

```python
# Encoding categorical variables
le = LabelEncoder()
label_columns = ['Status', 'Suburb', 'CouncilArea', 'Regionname', 'seasons']

for cols in label_columns:
    X[cols] = le.fit_transform(X[cols])

X.Type.replace({'h':1, 'u':2,'t':3}, inplace = True)
```

### Splitting Data

```python
# Splitting data into train, validation and test data
x_tr, x_test, y_tr, y_test = train_test_split(X, y, train_size = 0.75, random_state = 42)
x_train, x_val, y_train, y_val = train_test_split(x_tr, y_tr, test_size = 0.2, random_state = 42)
```

### Model Selection

```python
def make_results(model_name: str, model_object, metric: str):
    '''
    Arguments:
        model_name (string): name for the model in output table
        model_object: a fit GridSearchCV object
        metric (string): 'rmse', 'mae', or 'r2'

    Returns:
        pandas df with RMSE, MAE, and RÂ² for the best model based on the chosen metric
    '''
    # Get CV results
    best_idx = model_object.best_index_
    cv_results = model_object.cv_results_

    # CV RMSE (average over folds for best params)
    cv_rmse = np.sqrt(-cv_results['mean_test_neg_mean_squared_error'][best_idx])
    r2 = cv_results['mean_test_r2'][best_idx]
    
    # Predict on test set
    y_pred_test = model_object.best_estimator_.predict(x_test)

    # Test RMSE
    test_rmse = mean_squared_error(y_test, y_pred_test, squared=False)
    test_r2 = r2_score(y_test, y_pred_test)

    dataframe_1 = pd.DataFrame({
        'model': [model_name],
        'RMSE': [cv_rmse],
        'R2': [r2] })

    dataframe_2 = pd.DataFrame({
        'model': ['best_' + model_name],
        'RMSE': [test_rmse],
        'R2': [test_r2] })
        
    result = pd.concat([dataframe_1, dataframe_2],axis = 0)
          
    return result
```

### Random Forest Regressor

```python
%%time
# Instantiate Random Forest Regressor
rf = RandomForestRegressor(random_state = 42)

# Defining RF hyperparameters for GridSearchCV 
cv_params = {'max_depth': [None],
             'max_features': [None],
             'max_samples': [1],
             'min_samples_leaf': [2],
             'min_samples_split': [2],
             'n_estimators': [200, 300],
             }

# Scoring metrics
scoring = ['r2', 'neg_mean_squared_error']

# Instantiate GridSearchCV for RF
rf_cv = GridSearchCV(rf, cv_params, scoring = scoring, cv = 3, refit = 'r2')

# Fitting model with GridSearchCV
rf_cv.fit(x_train, y_train)
```

```python
# Calling make_results function
result = make_results('RF', rf_cv, 'r2')
result
```

### Decision Tree Regressor

```python
%%time
# Instantiate Decision Tree Regressor
dtr = DecisionTreeRegressor(random_state = 42)

# Defining DTR hyperparameters for GridSearchCV 
cv_params = {'max_depth': [None],
             'max_features': [None],
             'min_samples_leaf': [4,6,8,10],
             'min_samples_split': [2,4,6]
             }

# Instantiate GridSearchCV for DTR
dtr_cv = GridSearchCV(dtr, cv_params, scoring = scoring, refit = 'r2')

# Fitting model with GridSearchCV
dtr_cv.fit(x_train, y_train)
```

```python
# Concatenate DTR results
dtr_result = make_results('DTR', dtr_cv, 'r2')
result_2 = pd.concat([result, dtr_result],axis = 0)
result_2.sort_values(by = 'R2', ascending = False)
```

### XGBoost Regressor

```python
%%time
# Instantiate XGBoost Regressor
xgb = XGBRegressor(random_state = 42)

# Defining XGB hyperparameters for GridSearchCV 
cv_params = {'max_depth': [10],
             'min_child_weight': [3],
             'learning_rate': [0.1,0.3],
             'n_estimators': [50, 100]
             }

# Instantiate GridSearchCV for XGB
xgb_cv = GridSearchCV(xgb, cv_params, scoring = scoring, refit = 'r2')

# Fitting model with GridSearchCV
xgb_cv.fit(x_train, y_train)
```

```python
# Concatenate XGB results
xgb_result = make_results('XGB', xgb_cv, 'r2')
result_3 = pd.concat([result_2, xgb_result],axis = 0)
result_3.reset_index().sort_values(by = 'R2', ascending = False)
```
